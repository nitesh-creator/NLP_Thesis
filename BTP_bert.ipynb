```python
train = pd.read_csv("/content/drive/MyDrive/BTP/ecommerceDataset.csv")
```

```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
train['Category']= label_encoder.fit_transform(train['Category'])
```

```python
from sentence_transformers import SentenceTransformer
sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
train
```

```python
sentences=np.array(train['Text'])
```

```python
sentence_embeddings = sbert_model.encode(sentences)
```

```python
sentence_embeddings
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test = train_test_split(sentence_embeddings,np.array(train['Category']),random_state=104,test_size=0.25,shuffle=True)
```

```python
clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```
